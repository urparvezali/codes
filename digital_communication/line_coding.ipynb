```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# np.random.seed(22)
data = [np.random.randint(0, 2) for _ in range(10)]
print(data)
```

# Unipolar non-return to zero

```python
from math import floor
unipolar_nonreturntozero = [1 if data[int(floor(x))] else 0 for x in np.arange(0, 10, 0.001)]
```

```python
plt.plot(unipolar_nonreturntozero)
plt.ylim(-1.5, 1.5)
```
